```python
FileName = "20181111140403" 
DataStream = read("./Data/%s.PICKLE" %FileName,format='PICKLE')
```

```python
cat = read_events('/raid1/Data/ChinaArray/20181111/CMTSOLUTION')
EventDepth = cat[0].origins[0].depth/1000.
EventLat, EventLat = cat[0].origins[0].latitude, cat[0].origins[0].longitude
EventTime = cat[0].origins[0].time
```

```python
print(DataStream.select(component="T")[0].stats.traveltimes['S'])
```

```python
## Frequencies for filter
fmin = 1/30. #Hz
fmax = 1/10.  #Hzseis

dist_range_0 = 100
dist_range_1 = 110
dist_range_2 = 120
dist_range_3 = 130
dist_range_4 = 140

azi_min = 0
azi_max = 360

time_min = -20
time_max = 3600

linewidth = 0.5
norm_constant = 1

if_round = True

switch_yaxis = True
```

```python
fig, ax = plt.subplots(1,4,sharey=True,sharex=True)
fig.set_size_inches(20,30)


DataStream.filter('bandpass', freqmin=fmin,freqmax=fmax, zerophase=True)
DataStream.differentiate()

norm = np.max(DataStream.select(component="T")[0].data) / norm_constant

azis=[]
dists=[]
for itrace, trace in enumerate(DataStream.select(component="T")):
    dist = trace.stats.distance_in_degrees
    azi = trace.stats.azimuth
#     print("dist %.1f azi %.1f" %(dist, azi))
    # Split seismograms by distance range (this needs to be adapted per Event to produce a reasonable plot.
    if dist < dist_range_1 and dist > dist_range_0:
        i_axis = 0
    elif dist < dist_range_2 and dist > dist_range_1:
        i_axis = 1
    elif dist < dist_range_3 and dist > dist_range_2:
        i_axis = 2
    elif dist < dist_range_4 and dist > dist_range_3:
        i_axis = 3
    else:
        continue

    try:
        phase_time = trace.stats.traveltimes['S'] or trace.stats.traveltimes['Sdiff']
        align_time = phase_time
    except:
        continue
    
    if if_round:
        azi = np.round(azi)    
    
    plot_time = trace.times(reftime=EventTime)-align_time
    plot_data = trace.data / np.max(trace.data) * norm_constant + azi
    ax[i_axis].plot(plot_time, plot_data,'k',linewidth=linewidth)
    
    azis.append(azi) # list all azimuths
    dists.append(dist)
    
azis = np.array(azis)
dists = np.array(dists)

ax[0].set_xlim(time_min,time_max)
ax[0].set_ylim(azis.min()-3,azis.max()+3)

ax[0].set_title(' Sdiff dist %d - %d' %(dist_range_0,dist_range_1))
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Azimuth [$^{\circ}$]')

ax[1].set_title(' Sdiff dist %d - %d' %(dist_range_1,dist_range_2))
ax[1].set_xlabel('Time [s]')

ax[2].set_title(' Sdiff dist %d - %d' %(dist_range_2,dist_range_3))
ax[2].set_xlabel('Time [s]')

ax[3].set_title(' Sdiff dist %d - %d' %(dist_range_3,dist_range_4))
ax[3].set_xlabel('Time [s]')

if switch_yaxis:
   ax[0].invert_yaxis()
```
